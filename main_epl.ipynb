```python
import pandas as pd

try:
    # Đọc dữ liệu từ các file CSV
    goalkeeper_df = pd.read_csv('goalkeeping.csv')
    standard_df = pd.read_csv('standard.csv')
    shooting_df = pd.read_csv('shooting.csv')
    passing_df = pd.read_csv('passing.csv')
    
    # Kiểm tra xem các DataFrame có chứa cột 'Id' hay không
    if 'Id' not in goalkeeper_df.columns or 'Id' not in standard_df.columns or 'Id' not in shooting_df.columns:
        raise KeyError("Một hoặc nhiều file không chứa cột 'Id'. Vui lòng kiểm tra lại dữ liệu.")
    
    # Merge các DataFrame theo cột 'Id'
    merged_df = pd.merge(standard_df, goalkeeper_df, on='Id', how='outer')
    merged_df = pd.merge(merged_df, shooting_df, on='Id', how='outer')
    merged_df = pd.merge(merged_df, passing_df, on='Id', how='outer')
    
    # Xử lý các giá trị null (tùy chọn)
    # merged_df.fillna(0, inplace=True)  # Ví dụ: điền giá trị 0 cho các ô bị thiếu
    
    # Lưu dữ liệu đã merge vào file CSV mới
    merged_df.to_csv('merged_goalkeeper_standard.csv', index=False)
    
    print("Merge hoàn thành! Dữ liệu đã được lưu vào 'merged_goalkeeper_standard.csv'.")
    
except FileNotFoundError as e:
    print(f"Lỗi: Không tìm thấy file - {e.filename}")
except KeyError as e:
    print(f"Lỗi: {e}")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")
```
